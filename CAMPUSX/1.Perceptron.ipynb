```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('placed_data.csv')
```

```python
print(df.shape)
df.head(10)
```

```python
sns.scatterplot(x=df['cgpa'],y=df['resume_score'],hue=df['placed'])
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
model = Perceptron()
```

```python
model.fit(X,y)
```

```python
#model.coef_ gives you the learned weights (coefficients) of your trained model.
model.coef_
```

```python
#model.intercept_ (note the underscore) gives you the bias term of the model.
model.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values,y.values,clf=model,legend=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

X = df[['cgpa', 'resume_score']].values
y = df['placed'].values

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train
model = LogisticRegression()
model.fit(X_scaled, y)

# Plot using scaled X
plt.figure(figsize=(8,6))
plot_decision_regions(X_scaled, y, clf=model, legend=2)
plt.xlabel("CGPA (scaled)")
plt.ylabel("Resume Score (scaled)")
plt.show()
```
