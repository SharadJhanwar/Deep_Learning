```python
import numpy as np

class Perceptron:
    def __init__(self, learning_rate=0.1, n_iterations=10):
        self.lr = learning_rate
        self.n_iters = n_iterations
        self.activation = lambda x: 1 if x >= 0 else 0

    def fit(self, X, y):
        # Initialize weights (w) and bias (b)
        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        for _ in range(self.n_iters):
            for x_i, target in zip(X, y):
                # Calculate prediction
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_pred = self.activation(linear_output)

                # Update rule for weights and bias
                update = self.lr * (target - y_pred)
                self.weights += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return np.array([self.activation(output) for output in linear_output])

# Helper function to test and report
def test_perceptron(X, y, label):
    model = Perceptron()
    model.fit(X, y)
    predictions = model.predict(X)
    accuracy = np.mean(predictions == y)
    print(f"--- {label} Test ---")
    print(f"Predictions: {predictions}")
    print(f"True Labels: {y}")
    print(f"Accuracy: {accuracy * 100:.2f}%")
    return accuracy
```

```python
# Data for the AND gate
X_and = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_and = np.array([0, 0, 0, 1])

# Run the test
test_perceptron(X_and, y_and, "AND Gate")
```

```python
# Data for the XOR gate
X_xor = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_xor = np.array([0, 1, 1, 0])

# Run the test
test_perceptron(X_xor, y_xor, "XOR Gate")
```
