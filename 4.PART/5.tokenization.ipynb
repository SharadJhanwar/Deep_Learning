### Tokenization 
means breaking raw text into smaller units called tokens so a machine learning model (or NLP system) can understand and process it.

In short:

Text → Tokens → Numbers → Model

### What is a Token?

A token can be:

A word

A subword

A character

A symbol

Depends on the tokenizer.

Examples
Tokenization type	Example
Word-level	"I love AI" → ["I", "love", "AI"]
Character-level	"AI" → ["A", "I"]
Subword-level (most common today)	"tokenization" → ["token", "ization"]
Byte-level	Used internally by GPT-style models

```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
text = "Time is time, care is care, tokenization is veryimportant"
```

```python
tokens = tokenizer.tokenize(text=text)
```

```python
ids = tokenizer(text)
ids
```

```python
text_a = "Tokenization is important"
text_b = "Models need attention"
```

```python
output = tokenizer(
    text = text_a,
    text_pair = text_b,
    padding=True,
    truncation=True,
    return_tensors=None
)

print(output)
```
